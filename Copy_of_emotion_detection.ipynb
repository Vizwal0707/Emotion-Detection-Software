```python
from deepface import DeepFace
import cv2
import matplotlib.pyplot as plt
img1=cv2.imread("/content/images.jpg")
plt.imshow(img1)
plt.show
```

```python
plt.imshow(img1[:, :, ::-1])
plt.show()
```

```python
obj=DeepFace.analyze(img1, actions = ['age', 'gender', 'race', 'emotion'])
print(obj)
```

```python
print(type(obj))
```

Convert it into Dictionary

```python
final_obj={}
for i in obj:
  final_obj.update(i)
  print(final_obj)
  print(type(final_obj))
```

```python
print(range(len(final_obj)))
print(final_obj['age'])
print(final_obj['dominant_gender'])
print(final_obj['dominant_race'])
print(final_obj['dominant_emotion'])
```

```python
print("Your age is",final_obj['age'],"and you are",final_obj['dominant_race'],"and current emotion is", final_obj['dominant_emotion'], "and your gender capture is", final_obj['dominant_gender'])
```
